```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , Input , Flatten , Conv2D , MaxPooling2D
```

```python
train_data = pd.read_csv('mnist_train.csv')
test_data = pd.read_csv('mnist_test.csv')
```

```python
train_data.head()
```

```python
28*28
```

```python
X_train = train_data.drop(columns=['label'])
y_train = train_data['label']
```

```python
test_data.head()
```

```python
X_test = test_data.drop(columns=['label'])
y_test = test_data['label']
```

```python
X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
X_train = X_train /255
X_test = X_test/255
```

```python
28*28
```

```python
X_train = X_train.to_numpy().reshape(60000,28,28)
X_test = X_test.to_numpy().reshape(10000,28,28)
```

```python
X_train.shape
```

```python
model = Sequential([
    Conv2D(28,(3,3),activation='relu'),
    MaxPooling2D(),
    Conv2D(56,(3,3),activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(128,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='sgd' , loss = 'sparse_categorical_crossentropy' , metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=11,validation_data=(X_test , y_test))
```

```python
plt.plot(history.history['loss'],label='Training Loss')
plt.plot(history.history['val_loss'],label='Testing Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.plot(history.history['accuracy'],label='Training Accuracy')
plt.plot(history.history['val_accuracy'],label='Testing Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
